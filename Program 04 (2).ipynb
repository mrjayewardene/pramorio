```python
def aretheyvalid(A, B, C): 
    if A >100:
        print(f'{A}, {B}, {C}')
        print(f'Invalid group -> {A} is too big')
    if A <0:
        print(f'{A}, {B}, {C}')
        print(f'Invalid group -> {A} is too small')
    if B >100:
        print(f'{A}, {B}, {C}')
        print(f'Invalid group -> {B} is too big')
    if B <0:
        print(f'{A}, {B}, {C}')
        print(f'Invalid group -> {B} is too small')
    if C >100:
        print(f'{A}, {B}, {C}')
        print(f'Invalid group -> {C} is too big')
    if C <0:
        print(f'{A}, {B}, {C}')
        print(f'Invalid group -> {C} is too small')
    X = (A<=100 and A>=0) and (B<=100 and B>=0) and (C<=100 and C>=0)
    if (X):
        return True
    else:
        return False
```
