```python
for index, row in df.iterrows():
    A = row['s1']
    B = row['s2']
    C = row['s3']
    def aretheyvalid(A, B, C):
        if A >100:
            print(f'{A} is too big')
        if A <0:
            print(f'{A} is too small')
        if B >100:
            print(f'{B} is too big')
        if B <0:
            print(f'{B} is too small')
        if C >100:
            print(f'{C} is too big')
        if C <0:
            print(f'{C} is too small')        
        if (A<=100 and A>=0) and (B<=100 and B>=0) and (C<=100 and C>=0):
            return True
        else:
            return False 
    aretheyvalid(A,B,C)
```
